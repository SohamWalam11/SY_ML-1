# Q2

  Load data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score

# Load the dataset
df = pd.read_csv('creditcard.csv')

# Show the first few rows
df.head()
```

Imbalance in Dataset

```python
# Check class distribution
class_counts = df['Class'].value_counts()

print("Class distribution:\n", class_counts)

# Visualize class distribution
sns.countplot(x='Class', data=df)
plt.title('Class Distribution')
plt.show()

# Calculate imbalance ratio
imbalance_ratio = class_counts[0] / class_counts[1]
print(f"Imbalance Ratio (Normal : Fraud) = {imbalance_ratio:.2f} : 1")
```

Data for Modelling


```python
# Define features and target
X = df.drop('Class', axis=1)
y = df['Class']

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

Train and evluate Decision tree

```python
# Initialize Decision Tree
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

# Predictions
y_pred_dt = dt.predict(X_test)

# Evaluation
print("Decision Tree Classifier Results:")
print(confusion_matrix(y_test, y_pred_dt))
print(classification_report(y_test, y_pred_dt))
print("ROC-AUC Score:", roc_auc_score(y_test, y_pred_dt))
print("Accuracy:", accuracy_score(y_test, y_pred_dt))
```

Train and evaluate AdaBoost Classifier

```python
# Initialize AdaBoost
adb = AdaBoostClassifier(random_state=42)
adb.fit(X_train, y_train)

# Predictions
y_pred_adb = adb.predict(X_test)

# Evaluation
print("AdaBoost Classifier Results:")
print(confusion_matrix(y_test, y_pred_adb))
print(classification_report(y_test, y_pred_adb))
print("ROC-AUC Score:", roc_auc_score(y_test, y_pred_adb))
print("Accuracy:", accuracy_score(y_test, y_pred_adb))
```

Analysis and Comparison

```python
print("Comparison Analysis:\n")
print("1. Decision Tree tends to overfit and may not handle imbalance well alone.")
print("2. AdaBoost generally performs better in imbalanced scenarios as it focuses on misclassified instances iteratively.")
print("3. Look at the recall for the '1' class (Fraud). Higher recall means fewer false negatives (which is critical here).")
print("4. ROC-AUC is a good indicator for imbalanced classes. Higher is better.")

print("\nRecommendation:")
print("- Since this is a fraud detection problem, focus on **Recall** and **ROC-AUC** for class 1 (fraud).")
print("- AdaBoost often outperforms a standalone Decision Tree in such scenarios.")
```
